
### Что такое Data Science?

**Данные** это серый кардинал. Сегодня данные правят миром, товары в рекомендациях, социальные опросы, лучший маршрут без пробок, прогноз котировок. Все они используют анализ данных.

**Data Science** - наука об изучении данных, для извлечения полезной информации при помощи различных научных методов, алгоритмов и процессов.


 


#### Зачем нам Data Science?

**Data Science** - наука изучающая данные, помогает принять наиболее правильные решения, обнаруживать закономерности, выполнять анализ и визуализировать данные. Эта наука необходима для:

* Задавать правильные вопросы
* Находить закономерности
* Визуализировать результат анализа 
* Наделять машины интеллектом
* Принимать более качественные и быстрые решения




#### Ключевые даты развития Data Science.

#### Настоящее и будущее Data Science 

Сегодня развитие Data Science идет по нескольким веткам.

* **`Искуственный интелект`** - сфера создания интеллектуальных машин, способных работать и принимать решения как человек. Исследования в этой области начались в 1936 г., Алан Тьюринг создал первые машины на базе ИИ.

* **`Машинное обучение`** - создание различных моделей, об**учение с учителем и** обучение без учителя.

* **`Глубокое обучение`** - создание многослойной нейросети, наличие более одного скрытого слоя нейронов в сети, которые проводят математические вычесления.

#### Основы компьютерного программирования

Программирование это общение с компьюетером, пошаговые инструкции, понятные компьютеру.

Пошаговые инструкции называются **алгортимом**.

**Алгоритм** - последовательность четко определенных действий. 

Чем проще и оптимальнее алгоритм тем лучше и быстрее результат.

Алгоритмы состоят из 3 видов операторов.

1. Последовательные операторы. 
2. Условные операторы. 
3. Циклы или итерации.

* **Последовательные операторы** - выполняются один за другим, последовательно.
* **Условные операторы** - выполняются в зависимости от определеннго условия.
* **Циклы или итерации** - определенный набор действий выполняется снова и снова в зависимости от некоторого условия.


##### Языки программирования

Существует огромное количество языков программирования, по аналогии с разговорными языками, у всех языков программирования есть свои сильные и слабые стороны в зависимости от задач.

Компьютер понимает только **машинный язык**, языки программирования созданы что бы человек мог взаимодействовать с машиной. 

Все языки программирования являются **языками выского уровня**


##### Исходный код

Исходный код - то что пишут программисты на всех языках программирования. Исходный код не может содержать разметки и форматирования. 

Что бы машина понимала исходный код необходимо перевести код на понятный для машины язык. Для этого в помощь нам приходят такие понятия как ***компиляция*** и ***интерпретация***

**Компилятор** - это прогармма, преобразующая исходный код в машинный. Комплияция языка программирования высокго уровня на машинный язык.

**Интерпретатор** - это программа которая преобразует исходный код в машинный при помощи построчного чтения кода, при каждом запуске исходного кода.

***Наш любимый Python использует комбинацию обоих методов.***
